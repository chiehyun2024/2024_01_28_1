```python
import matplotlib.pyplot as plt
# 1.製作figure ->畫布
fig = plt.figure()
print(type(fig))

# 2.在1.的figure上製作一個axes ->圖表
ax_1 = fig.add_subplot(1, 1, 1)
print(type(ax_1))
ax_1.set_title('Ax') #圖表名稱

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
days = [num for num in range(1,32)]
celsius = [round( random.uniform(28.5,35.2),ndigits=1) for _ in range(1,32)]
celsius1 = [round( random.uniform(28.5,35.2),ndigits=1) for _ in range(1,32)]
figure = plt.figure(figsize=(10,5))  #figsize表格寬度
ax = figure.add_subplot(1,1,1)
ax.grid() #表格線
ax.plot(days,celsius,'g.-.') #X軸.Y軸.顏色線條
ax.plot(days,celsius1,'r-')
ax.set_title('july and august weather')
ax.set_xlabel('Days')
ax.set_ylabel('celsius')
ax.legend(['july','august']) #線條名稱
#print(ax.get_xticks())  #顯示本來用的日期
ax.set_xticks(days) #用自己的日期
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
t = np.arange(0,2.5,0.1)  #數字範圍在0-2.5,相差0.1
y1 = np.sin(t * math.pi) #使用了數學運算
y2 = np.cos(t * math.pi)
#y2 = np.sin(t * math.pi + math.pi/2)

figure = plt.figure()
ax = figure.add_subplot(1,1,1)
ax.plot(t,y1,'b*')
ax.plot(t,y2,'g^')
plt.show()
```
