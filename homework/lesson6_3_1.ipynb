```python
#BMI值計算公式:    BMI = 體重(公斤) / 身高2(公尺2)

#檢查輸入的範圍
#使用pyinputplus
#輸出錯誤訊息

#輸入:
#請輸入身高(公分,120-250):
#請輸入體重(公斤,30-250):

#輸出:
#您的BMI是:xxxx

import pyinputplus as pyip

height = pyip.inputInt('身高:')
print(height)
if height >= 120 and height <= 250:
    print('在120-250之間')
else:
    print('不在範圍內') 

weight = pyip.inputInt('體重:')
print(weight)
if weight >= 30 and weight <= 250:
    print('在30-250之間')
else:
    print('不在範圍內')
bmi = weight/(height/100)**2
print(f'您的BMI是:{bmi:.2f}')

if bmi<18.5:
    print('您的體重:體重過輕')
elif bmi<24:
    print('您的體重:正常')
elif bmi<27:
    print('您的體重:過重')
elif bmi<30:
    print('您的體重:輕度肥胖')
elif bmi<35:
    print('您的體重:中度肥胖')
else:
    print('您的體重:重度肥胖')
```
