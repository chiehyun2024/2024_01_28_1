```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
df1 = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51))
df1.index.name='學號'
df1.columns.name="科目"
df1
```

```python
from pandas import Series
def fun1(value:Series)->Series:
    return pd.Series([value.max(),value.min(),value.median()],['最高分','最低分','中間分'])

df2 = df1.apply(fun1)
df2
```

```python
#轉換成numpy
df1.values[:,0]
```

```python
#轉成csv
df2.to_csv('3年1班統計資料.csv')
```
